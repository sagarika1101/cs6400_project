```python
copy_complete_df = complete_df.copy()
copy_complete_df.head(5)
```

```python
newComplaint_df = pd.DataFrame(newComplaintNum, columns=['CMPLNT_NUM'])
newComplaint_df
```

```python
random_subset = copy_complete_df.sample(n=100)
random_subset = random_subset.reset_index()
```

```python
random_subset.drop(['index', 'CMPLNT_NUM'], axis=1, inplace=True)
```

```python
random_subset
```

```python
merged = newComplaint_df.merge(random_subset, left_index=True, right_index=True, how='inner')
```

```python
merged.to_csv('bogus_complete_data.csv', index=False)
```
